```python
import urllib.request
import os
def find_next_file(content):
    mouse_id = content.find('Mouse')
    mat_id = content.find('.mat')
    content = content[mat_id+4:]
    return content,content[mouse_id:mat_id+4]

def get_files(page):
    page = urllib.request.urlopen('https://wustl.app.box.com/s/shek70aq3jvil1vrku0xshv2cwai3dum'+page)
    content = str(page.read())
    files = []
    while 'Mouse' in content:
        content,nextfile = find_next_file(content)
        files.append(nextfile)
    files = [i for i in files if 'Mouse' in i]
    return files
```

```python
pages = [f'?page={i+1}' for i in range(25)]
pages.append('')
all_files = []
all_pages = []
for page in pages:
    files = get_files(page)
    all_files = all_files+files
    all_pages = all_pages+ [page for _ in range(len(files))]
all_files
```
